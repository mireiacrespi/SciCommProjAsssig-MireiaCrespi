```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Fri Mar  4 12:30:58 2022

@author: mireia
"""

import random #servirà per fer que la màquina trii nombres a l'atzar
taulell = [] #definim una matriu anomenada taulell (com a variable global)

# Funció: crear el taulell en forma de llista
# Paràmetres: taulell (variable global)
# Resultat: taulell (matriu)
 
def crear_taulell ():
    for fila in range (6):
        taulell.append ([])
        for columna in range (7):
            taulell[fila].append ("_") #cada forat de la matriu té _
    
    return taulell



# Funció: ensenyar el taulell en forma de matriu
# Paràmetres: taulell (variable global)

def ensenyar_taulell ():
    for fila in range (6):
        print ("|", end = " ") #perquè al principi de cada fila hi hagi |
        for columna in range (7):
            print (taulell[fila][columna], end = " ")
            print ("|", end = " ") #perquè cada espai del taulell estigui separat per |
        print (" ") #canviar de fila
    
    return



# Funció: el jugador tria si vol jugar amb la màquina o amb algú altre
# Paràmetres: taulell (variable global)
# Resultat: joc (booleà)

def mode_joc ():
    c = False #per regular el bucle

    while c == False :
        pregunta = int (input ("Escriu 1 si vols jugar amb la màquina o 2 si vols jugar amb un amic:  "))
        if pregunta == 1 :
            joc = True #si joc és True, vol jugar amb la maquina
            c = True
        elif pregunta == 2 :
            joc = False #si joc és False, vol jugar amb una persona
            c = True
    #Si no passa cap d'aquestes dues coses, no sortirà del bucle, tornarà a demanar un mode de joc
    
    return joc



# Funció: el jugador 1 tria a quina columna vol posar la seva fitxa
# Paràmetres: taulell (variable global)

def tirada_jugador1 (): 
    
    col = int (input ("Jugador 1, tria la columna on vols tirar la teva fitxa (X) : "))
    columna = col - 1 #per començar a contar desde 0   
   
    f = False #suposem que sempre ens donarà una columna incorrecte per comprovar-ho sempre
    
   
    
    
    while f == False:
        if columna > 6 : #per evitar que el jugador seleccioni una columna inexistent (index out of range)
            print ("Columna inexistent")
            col = int (input ("Selecciona una altra columna: "))
            columna = col - 1 
            
        elif taulell[0][columna] != "_": #per evitar posar fitxes en columnes plenes
            print ("Columna plena")
            col = int (input ("Selecciona una altra columna: "))
            columna = col - 1 
      
        else :
            f = True #per sortir del bucle
    

    fila = 5 #comencm per la última fila
    buit = True #suposem que la posició seleccionada està buida
    
    while fila >= 0 and buit == True:
        if taulell[fila][columna] == "_" :
            taulell[fila][columna] = "X"  
            buit = False #per sortir del bucle
        else :
            fila = fila - 1
               
    return 
        
 

# Funció: el jugador 2 tria a quina columna vol posar la seva fitxa (O)
# Paràmetres: taulell (variable global)

def tirada_jugador2 (): 

    col = int (input ("Jugador 2, tria la columna on vols tirar la teva fitxa (O) : "))
    columna = col - 1 #per començar a contar desde 0 

    f = False 
    
    while f == False : 
        if columna > 6 : #per evitar que el jugador seleccioni una columna inexistent (index out of range)
            print ("Columna inexistent")
            col = int (input ("Selecciona una altra columna: "))
            columna = col - 1
            
        elif taulell[0][columna] != "_":
            print ("Columna plena")
            col = int (input ("Selecciona una altra columna: "))
            columna = col - 1 
        
       
        
        else :
            f = True #surt del bucle
    
    fila = 5 #comencem per la última fila
    buit = True
    
    while fila >= 0 and buit == True :
        if taulell[fila][columna] == "_" :
            taulell[fila][columna] = "O"  
            buit = False #perque no torni a entrar al bucle
        else :
            fila = fila - 1
               
    return



# Funció: tirada aleatoria de la maquina (O)
# Paràmetres: taulell (variable global)

def tirada_maquina (): 
    columna = random.randint (0, 6) #la columna és un número aleatori
    fila = 5 #comencem per la última fila
    buit = True
    
    while fila >= 0 and buit == True :
        if taulell[fila][columna] == "_" :
            taulell[fila][columna] = "O"
            buit = False
        
        elif taulell[0][columna] != "_": #entrarà en el bucle si la primera posició d'una columna no està buida
            print ("Columna plena, la màquina tria una altra columna")
            columna = random.randint (0, 6) #torna a donar un nombre aleatori
        
        else :
            fila = fila - 1
    
    return



# Funció: comprovar si hi ha un quatre en ratlla vertical
# Paràmetres: taulell (variable global)
# Resultat: ratlla4_v (booleà)

def comprovar_vertical ():
    columna = 0 #comença per la primera columna
    ratlla4_v = False #per començar, suposem que no hi ha 4 en ratlla
    
    while columna < 6 :
        fila = 0 #comença per la primera fila
        while fila < 3 : #només fins la fila 3 perquè no es poden fer més ratlles de 4 en sis espais
            if taulell[fila][columna] == taulell[fila + 1][columna] and taulell[fila + 1][columna] == taulell[fila + 2][columna] and taulell[fila + 2][columna] == taulell[fila + 3][columna] and taulell[fila][columna] != "_"  :
                #hi ha 4 posicions iguals seguides augmentant la fila, per tant estan totes en la mateixa columna
                ratlla4_v = True
                
            fila = fila + 1 #per repassar totes les files fins les 3
        columna = columna + 1 #per repassar totes les columnes

    return ratlla4_v



# Funció comprovar si hi ha quatre en ratlla horizontal
# Paràmetres: taulell (variable global)
# Resultat: ratlla4_h (booleà)

def comprovar_horizontal ():
    fila = 0
    ratlla4_h = False
    
    while fila < 6 :
        columna = 0
        while columna < 4 : #només fins la fila 4 perquè no es poden fer més ratlles de 4 en set espais
            if taulell[fila][columna] == taulell[fila][columna + 1] and taulell[fila][columna + 1] == taulell[fila][columna + 2] and taulell[fila][columna + 2] == taulell[fila][columna + 3] and taulell[fila][columna] != "_"  :
                #hi ha 4 posicions iguals seguides augmentant la columna, per tant estan totes en la mateixa fila
                ratlla4_h = True
                
            columna = columna + 1
        fila = fila + 1
    
    return ratlla4_h



# Funció: comprovar si hi ha quatre en ratlla diagonal positiva
# Paràmetres: taulell (variable global)
# Resultat: ratlla4_d1 (booleà)

def comprovar_diagonal1 () :
    fila = 0
    ratlla4_d1 = False
    
    while fila < 3 : #només fins la fila 3 perquè en la cantonada de baix esquerra no es pot fer una diagonal positiva de 4
        columna = 0
        while columna < 4 : #només fins la columa 4 perquè en la cantonada de dalt dreta no es pot fer una diagonal positiva de 4
            if taulell[fila][columna] == taulell[fila + 1][columna + 1] and taulell[fila + 1][columna + 1] == taulell[fila + 2][columna + 2] and taulell[fila + 2][columna + 2] == taulell[fila + 3][columna + 3] and taulell[fila][columna] != "_" :
                #hi ha 4 posicions iguals seguides augmentant la fila i la columna
                ratlla4_d1 = True
                
            columna = columna + 1
        fila = fila + 1
                
    return ratlla4_d1



# Funció: comprovar si hi ha quatre en ratlla diagonal negativa
# Paràmetres: taulell (variable global)
# Resultat: ratlla4_d2 (booleà)

def comprovar_diagonal2 () :
    fila = 0
    ratlla4_d2 = False
    
    while fila < 3 : #només fins la fila 3 perquè en la cantonada de baix dreta no es pot fer una diagonal negativa de 4
        columna = 6
        while columna > 2 : #la columna ha de ser més gran que 2 perquè en la cantonada de dalt esquerra no es ot fer una diagonal negativa de 4
            if taulell[fila][columna] == taulell[fila + 1][columna - 1] and taulell[fila + 1][columna - 1] == taulell[fila + 2][columna - 2] and taulell[fila + 2][columna - 2] == taulell[fila + 3][columna - 3] and taulell[fila][columna] != "_" :
                #hi ha 4 posicions iguals seguides disminuint la fila i la columna
                ratlla4_d2 = True
            
            columna = columna - 1
        fila = fila + 1
    
    return ratlla4_d2
   
 

# Funció: mirar si el taulell està ple
# Paràmetres: taulell (variable global)
# Resultat: empat (booleà)

def empat ():
    empat = True #suposem que està tot el taulell ple
    for fila in range (6) :
        for columna in range (7) :
            if taulell[fila][columna] == "_" :
                empat = False #si hi ha algun espai "buit" en tota la matriu, llavors el taulell no està ple
       
    return empat




# Mecànica del joc (es crida les funcions creades)
def main ():
    taulell = crear_taulell ()

    joc = mode_joc ()
  

    if joc == True : #juga amb la màquina
        print ("Has seleccionat jugar contra la màquina.")
        ensenyar_taulell ()      
        taulell_ple = empat ()

        a = False #paràmetre per controlar el bucle
        while empat() == False and a == False :
            tirada_jugador1 ()
            ensenyar_taulell()

            if comprovar_horizontal() == True or comprovar_vertical() == True or comprovar_diagonal1() == True or comprovar_diagonal2() == True :
                #comprova si hi ha 4 en ratlla
                print ("HAS GUANYAT!")
                a = True #surt del bucle
                
            else :
                print ("És el torn de la màquina:")
                tirada_maquina ()
                ensenyar_taulell ()

                if comprovar_horizontal() == True or comprovar_vertical() == True or comprovar_diagonal1() == True or comprovar_diagonal2() == True :
                    print ("Ha guanyat la màquina")
                    a = True # perque surti del bucle

                elif empat () == True :
                    print ("S'ha acabat la partida, heu empatat")
            
 
    
    else : #juga amb una persona
        print ("Has seleccionat jugar contra un altre jugador.")
        ensenyar_taulell ()      
        taulell_ple = empat ()

        a = False
        while a == False :
            tirada_jugador1 ()
            ensenyar_taulell()

            if comprovar_horizontal() == True or comprovar_vertical() == True or comprovar_diagonal1() == True or comprovar_diagonal2() == True :
                print ("HA GUANYAT EL JUGADOR 1!")
                a = True #perque surti del bucle

            else :
                tirada_jugador2 ()
                ensenyar_taulell ()

                if comprovar_horizontal() == True or comprovar_vertical() == True or comprovar_diagonal1() == True or comprovar_diagonal2() == True :
                    print ("HA GUANYAT EL JUGADOR 2!")
                    a = True # perque surti del bucle

                elif empat() == True :
                    print ("S'ha acabat la partida, heu empatat")
                    a = True


if __name__ == "__main__":
        main ()
        
   
```
